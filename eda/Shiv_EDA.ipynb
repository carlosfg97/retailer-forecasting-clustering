A transferred day is more like a normal day than a holiday. To find the day that it was actually celebrated, look for the corresponding row where type is Transfer

## Mapping Locale Names to cities/regions/country

There are 24 locale names out of which there are:
- 19 cities
- 4 regions
- 1 country

```python
events.groupby(['locale','locale_name']).count()
```

```python
cities = list(events.loc[events['locale']=='Local']['locale_name'].unique())
regions = list(events.loc[events['locale']=='Regional']['locale_name'].unique())
countries = ['Ecuador']
```

```python
def get_region_type_from_locale_name(locale_name):
    if locale_name in cities:
        return 'City'
    elif locale_name in regions:
        return 'Region'
    elif locale_name in countries:
        return 'Country'
    else:
        return None

events['region_type'] = events['locale_name'].apply(lambda x: get_region_type_from_locale_name(x))
events
```

```python
events['region_type'].value_counts().plot(kind='barh')
```

```python
national_events = events.loc[events['locale']=='National']
regional_events = events.loc[events['locale']=='Regional']
local_events = events.loc[events['locale']=='Local']
```

```python
local_events_holiday = local_events.loc[local_events['type']=='Holiday'][['date', 'locale_name']].drop_duplicates()
local_events_holiday['is_city_holiday'] = 1
local_events_holiday
```

## Joining transactions, stores, and events
1. Join transactions and stores on store_nbr
1. Join stores and local_events on stores.city = local_events.locale_name

```python
# Join Transactions and Stores on Store_Nbr
trans_stores = pd.merge(
    left=trans,
    right=stores,
    left_on='store_nbr',
    right_on='store_nbr',
    how='left'
)
trans_stores
```

```python
# Join trans_stores and local_events 
# on trans_stores.city = local_events.locale_name
# and trans_stores.date = local_events.date
trans_stores_events = pd.merge(
    left=trans_stores,
    right=local_events_holiday,
    left_on=['city','date'],
    right_on=['locale_name','date'],
    how='left'
)
# Drop locale name because it is the same as city
trans_stores_events.drop(columns='locale_name', inplace=True)

# Fillna with 0's
trans_stores_events['is_city_holiday'] = trans_stores_events['is_city_holiday'].fillna(0)

trans_stores_events
```

```python
trans_stores_events.groupby('city').sum()['transactions'].sort_values().plot(
    kind='barh',
    title='Transactions by City'
)
```

```python
trans_stores_events.groupby('state').sum()['transactions'].sort_values().plot(
    kind='barh',
    title='Transactions by State'
)
```

```python
pd.pivot_table(
    trans_stores_events,
    values='transactions',
    index='date',
    columns='type',
    aggfunc=np.sum).plot(figsize=(14, 8), title='Transactions over time by store type')
```

```python
pd.pivot_table(
    trans_stores_events,
    values='transactions',
    index='date',
    columns='state',
    aggfunc=np.sum).plot(figsize=(14, 8), title='Transactions over time by state')
```

```python
pd.pivot_table(
    trans_stores_events,
    values='transactions',
    index='date',
    columns='cluster',
    aggfunc=np.sum).plot(figsize=(14, 8), title='Transactions over time by cluster')
```
