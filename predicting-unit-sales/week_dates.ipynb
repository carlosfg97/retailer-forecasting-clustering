```python
import pandas as pd
```

```python
data_path = './data'
```

```python
cal = pd.read_csv(f'{data_path}/calendar_table.csv')
cal['date'] = pd.to_datetime(cal['date'])
grouper = pd.Grouper(key='date', freq='1W')
week_dates = pd.Series(cal.groupby(grouper).count()['day'].index)
week_dates_df = pd.DataFrame()
week_dates_df['week_date'] = pd.to_datetime(week_dates) + pd.Timedelta(days=1)
week_dates.to_csv(f'{data_path}/week_dates.csv', index=False)
week_dates_df
```
